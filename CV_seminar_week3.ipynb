# 데이터 로드

```python
dolphin_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/original/dolphin/*')
shark_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/original/shark/*')
whale_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/original/whale/*')
```

## 과제1번(9개의 사진)

```python
fig = plt.figure(figsize=(10,10))

for dolphin, shark, whale, ind in zip(dolphin_img_list, shark_img_list, whale_img_list, range(0, 3)):

  dolphin_ind = (ind*3) + 1
  dolphin = cv2.imread(dolphin)
  plt.subplot(3,3,dolphin_ind)
  if ind == 0 :
    plt.gca().set_title('dolphin')
  plt.imshow(dolphin[:,:,::-1]) 

  shark_ind = (ind*3) + 2
  shark = cv2.imread(shark)
  plt.subplot(3,3,shark_ind)
  if ind == 0 :
    plt.gca().set_title('shark')
  plt.imshow(shark[:,:,::-1])

  whale_ind = (ind*3) + 3
  whale = cv2.imread(whale)
  plt.subplot(3,3,whale_ind)
  if ind == 0 :
    plt.gca().set_title('whale')
  plt.imshow(whale[:,:,::-1])

```

## 과제2번(1개사진위에 5개의 선)

```python
import math
shark_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/original/shark/*')
image_1 = cv2.imread(shark_img_list[18])

height_up = math.ceil(((image_1.shape[0] / 5) / 2) - 5)
height_down = math.ceil(((image_1.shape[0] / 5) / 2) + 5)
weight_start = math.ceil((image_1.shape[1] / 10))
weight_end = math.ceil(image_1.shape[1] - (image_1.shape[1] / 10))
print('height_up : ', height_up,' height_donw : ', height_down)
print('weight_start : ', weight_start,' weight_end : ', weight_end)
```

```python
shark_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/original/shark/*')
```

```python
color_list = [(0,0,0),(0,0,255),(0,255,0),(255,0,0),(255,255,255)] 
image_1 = cv2.imread(shark_img_list[18])
for ii in range(1, 6):

  image_1[height_up+((ii-1)*55):height_down+((ii-1)*55), weight_start:weight_end,0] = color_list[ii-1][0]
  image_1[height_up+((ii-1)*55):height_down+((ii-1)*55), weight_start:weight_end,1] = color_list[ii-1][1]
  image_1[height_up+((ii-1)*55):height_down+((ii-1)*55), weight_start:weight_end,2] = color_list[ii-1][2]

cv2_imshow(image_1)
```

```python
for ii in range(1, 6):
  print('{}번째 색깔 조합 : '.format(ii), color_list[ii-1])
  print('{}번째 height : {} : {}'.format(ii, height_up+((ii-1)*55), height_down+((ii-1)*55)))
  print('{}번째 weight : {} : {}'.format(ii, weight_start, weight_end))
  print('\n')
```
